```python
#importing libraries 
import numpy as np
from matplotlib import pyplot as plt
```

```python
#calculating mean square error
def MSE(y,y_out):
    sum=0
    error=y-y_out
    for i in error:
        sum=sum+i*i
    return sum/len(y)
```

```python
#building the feature matrix X
def feature_matrix(x,m):
    X = np.empty((0,m+1),float)
    N = len(x)
    for i in range(N):
        temp1 = np.array([])
        for j in range(m+1):
            temp1 = np.append(temp1,pow(x[i],j))  
        X = np.append(X,np.array([temp1]),axis=0)
    return X
```

```python
#solving for the coefficients/weights
def solve(X,y):
    b = np.linalg.inv(X.T@X)@X.T@y
    return b
```

```python
#for plotting the datasets
def plot_data(x,y,y_out,m):
    plt.scatter(x, y,label = 'data_points')
    plt.scatter(x,y_out,color = 'r',label = 'predicted_data_points')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.show()
```

```python
# Generating training samples
#N = 10
N = 10
# Generate equispaced floats in the interval [0, 2π]
x = np.linspace(0, 2*np.pi, N)
# Generate noise
mean = 0
std = 0.05
# Generate some numbers from the sine function
y = np.sin(x)
# Add noise
y += np.random.normal(mean, std, N)
```

```python
#Generating test samples
T = 200
# Generate equispaced floats in the interval [0, 2π]
x_test = np.linspace(0, 2*np.pi, T)
# Generate noise
mean = 0
std = 0.05
# Generate some numbers from the sine function
y_test = np.sin(x_test)
# Add noise
y_test += np.random.normal(mean, std, T)
```

```python
#case(a) 
m = 1
X = feature_matrix(x,m)
b = solve(X,y)
X_test = feature_matrix(x_test,m)
y_out = b@X_test.T
plot_data(x_test,y_test,y_out,m)
print("Mean Square error : ",MSE(y_test,y_out))
print("This model is an example for underfitting")
```

```python
#case(b)
m = 2
X = feature_matrix(x,m)
b = solve(X,y)
X_test = feature_matrix(x_test,m)
y_out = b@X_test.T
plot_data(x_test,y_test,y_out,m)
print("Mean Square error : ",MSE(y_test,y_out))
print("This model is an example for underfitting")
```

```python
m = 3
X = feature_matrix(x,m)
b = solve(X,y)
X_test = feature_matrix(x_test,m)
y_out = b@X_test.T
plot_data(x_test,y_test,y_out,m)
print("Mean Square error : ",MSE(y_test,y_out))
print("This model is an example for appropriate-fitting")
```
