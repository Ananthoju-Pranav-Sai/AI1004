```python
#importing libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

#reading the data from train.csv
df = pd.read_csv('train.csv')

# Seperating the Labels with features
Y = df['SalePrice']

print(df.dtypes)
df
```

```python
# Dropping the salesprice as it is a label
df = df.drop('SalePrice',axis=1)

# Dropping the Id column as it is not considered as feature
df = df.drop('Id',axis=1)



#making the columns whose data-type is object to int type by converting them to category d-type and then assigning codes (in dictionary order)
for column in df:
    if(df[column].dtype=='object'):
        df[column] = df[column].astype("category").cat.codes

#Filling NA's if there are any with zeros
df.fillna(0,inplace=True)

#converting the dataframes to np.arrays as LinearRegressor() accepts np.arrays only
X_train = np.array(df) 
Y_train = np.array(Y)

#using LinearRegression() model to fit the data
model = LinearRegression()
model.fit(X_train,Y_train)

test = pd.read_csv('test.csv')

#storing the Id's in an different array
id = test['Id']
id = np.array([id])

# Dropping the Id column as it is not considered as feature
test = test.drop('Id',axis=1)

#making the columns whose data-type is object to int type by converting them to category d-type and then assigning codes (in dictionary order)
for column in test:
    if(test[column].dtype=='object'):
        test[column] = test[column].astype("category").cat.codes

#Filling NA's if there are any with zeros
test.fillna(0,inplace=True)
X_test = np.array(test)

predictions = model.predict(X_test)
values = np.append(id,np.array([predictions]),axis=0)

#creating a dataframe with Id's and predicted SalePrice of the test data
ans = pd.DataFrame({'Id': values[0, :], 'SalePrice': values[1, :]})

# As the sample_submission.csv has Id as index we will make Id as the index
ans =  ans.set_index('Id')

#Writing the dataframe to a csv
ans.to_csv('submission.csv')
ans
```
